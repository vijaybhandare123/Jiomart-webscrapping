```python
from bs4 import BeautifulSoup
import requests
import html5lib
import re
```

```python
url = 'https://www.jiomart.com'
```

```python
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text , 'html5lib' )
```

```python
menu = soup.find_all('li',{'class':'o-menu'} )
```

```python
sub_category=[]
```

```python
for i in menu:
    print(i.a['href'].split('/')[-2])
```

```python
sub_urls = soup.find_all(href = re.compile('fruits-vegetables'))
for i in sub_urls[1:-4]:
    
    sub_category.append(i['href'])
```

```python
sub_urls = soup.find_all(href = re.compile('dairy-bakery'))
for i in sub_urls[1:-5]:
    
    sub_category.append(i['href'])
```

```python
sub_urls = soup.find_all(href = re.compile('staples'))
for i in sub_urls[1:-9]:
    
    sub_category.append(i['href'])
```

```python
sub_urls = soup.find_all(href = re.compile('snacks-branded-foods'))
for i in sub_urls[1:-13]:
    
    sub_category.append(i['href'])
```

```python
sub_urls = soup.find_all(href = re.compile('beverages'))
for i in sub_urls[1:-5]:
    
    sub_category.append(i['href'])
```

```python
sub_urls = soup.find_all(href = re.compile('personal-care'))
for i in sub_urls[1:-13]:
    
    sub_category.append(i['href'])
```

```python
sub_urls = soup.find_all(href = re.compile('home-care'))
for i in sub_urls[1:5]:
    
    sub_category.append(i['href'])
    
```

```python
sub_urls = soup.find_all(href = re.compile('baby-care'))
for i in sub_urls[1:-4]:
    
    sub_category.append(i['href'])
```

```python
url4 = 'https://www.jiomart.com/offers'
response4 = requests.get(url4)
soup4 = BeautifulSoup(response4.text , 'html5lib' )
offers = soup4.find_all('span',{'class':'offer-coupon'} )[:3]
offer_detail=[]
for i in offers:
    offer_detail.append(i.text)
```

```python
Image = []
food_type_detail =[]
product_title = []
price_detail = []
stock_detail = []
seller_detail = []
disclaimer =[]
prod_description=[]
feature_detail=[]
country_of_origin = []
rating_detail =[]
tax=[]
offer_detail2 = []
product_div = []
you_save=[]
brand = []
manufacurer =[]
```

```python
sub_category
```

```python
for i in range(len(sub_category)):
    
    url2 = sub_category[i]

    while(1):
        
        



        response2 = requests.get(url2)
        soup2 = BeautifulSoup(response2.text , 'html5lib' )
        
        
        prod_div = soup2.find_all('title' )
        if prod_div==[]:
            main_div = 'NA'
        else:
            
            for i in prod_div:
                main_div = i.text[:-9]

        
        
        
        
        item = soup2.find_all('div',{'class':'cat-item'} )
        for j in item:
            url3 ='https://www.jiomart.com' + j.a['href']

            response3 = requests.get(url3)
            soup3 = BeautifulSoup(response3.text , 'html5lib' )
            
            ### 1) Main Product Division
            
            product_div.append(main_div)
               
                        
            ### 2) Product Name

            title = soup3.find_all('div',{'class':'title-section'} )
            if title ==[]:
                product_title.append('NA')
            else:    
                
                for i in title:
                    product_title.append(i.text)

            
        
            ## 1) Image of the product 

            product = soup3.find_all('div',{'class':'new_product'} )
            if product ==[]:
                Image.append('NA')
            else:
                
                for k in product:
                    Image.append(k.img['src'])

            ### 2) Food Type  - Veg or Non Veg

            food_type = soup3.find_all('span',{'class':'food_type_sec'} )
            
            if food_type == []:
                food_type_detail.append('NA')
            
            else:
            
                for i in food_type:
                    food_type_detail.append(i.span['class'][0])


            

            ### 4) Price of Product

            price = soup3.find_all('span',{'class':'final-price'} )
            
            if price == []:
                price_detail.append('NA')
                
            else:
                
                for i in price:
                    price_detail.append(i.text)
                    

            ### - 5) Stock Details

            stock = soup3.find_all('div',{'class':'stock_details'} )
            
            if stock == []:
                stock_detail.append('NA')
            
            else:
                
                for i in stock:
                    stock_detail.append(i.text)

                    
            ### - 6) Seller Details

            seller = soup3.find_all('div',{'class':'seller_details'} )
            if seller ==[]:
                seller_detail.append('NA')
            else:
                
                for i in seller:
                    seller_detail.append(i.text)
            
            
            
            ## 7)  Disclaimer Information

            drug = soup3.find_all('div',{'class':'drug-content'} )
            if drug==[]:
                disclaimer.append('NA')
            else:
                
                for i in drug:
                    disclaimer.append(i.text)

            ### 8) Description of Product

            para = soup3.find_all('p' )[3]
            if para == []:
                prod_description.append('NA')
                
            else:    
                for i in para:
                    prod_description.append(i)

            ### 9) Feature Details 

            features = soup3.find_all('div', {'class':'content_txt', 'id' : 'third_desc'})
            if features==[]:
                feature_detail.append('NA')
                
            else:                            
            
                for i in features:
                    feature_detail.append(i.text)
            
                  

            '''
            
            ### 11)  Ratings

            rating = soup3.find_all('div',{'class':'rating-content'} )
            for i in rating:
                rating_detail.append(i.text[1:4])
            '''
            #### - 12)  Tax Details 

            tax_detail = soup3.find_all('span',{'class':'tax_text'} )
            if tax_detail == []:
                tax.append('NA')
            
            else:
                
                for i in tax_detail:
                    tax.append(i.text)
                
            #### 13 Your Savings
            
            saving = soup3.find_all('span' , {'class' : 'disc-price'} )
            if saving == []:
                you_save.append('NA')
            
            else:
                
                for i in saving:
                    you_save.append(i.span.text)    

            #### -- Brand , Manufacture and country of origin 
            
            brand_manufacturer_country = soup3.find_all('td')
            
            if len(brand_manufacturer_country) == 4:
                
                brand1 = brand_manufacturer_country[0]
                manufacturer1 = brand_manufacturer_country[1]
                country1 = brand_manufacturer_country[2]
                brand.append(brand1.text)    
                manufacurer.append(manufacturer1.text)
                country_of_origin.append(country1.text)
                   
                
                        
            elif len(brand_manufacturer_country) == 2:  
                country1 = brand_manufacturer_country[0]
                brand1 = 'NA'
                manufacturer1 = 'NA'
                brand.append(brand1)    
                manufacurer.append(manufacturer1)
                country_of_origin.append(country1.text)
                
            else:
                
                brand.append('NA')    
                manufacurer.append('NA')
                country_of_origin.append('NA')
                
                
                       
            
                
                
            
            ####  13)  Offers 

            offer_detail2.append (offer_detail)
            
        
        next_page = soup2.find_all('li',{'class':'next'} )

        if next_page == [] :
            break

        else:
            url2 = [i.a['href'] for i in next_page]

            url2 = url2[0]   

```

```python
data = {"Product Division" :product_div ,"Product" : product_title , "Image" : Image   , "Price" : price_detail , "You Save" : you_save , 
         "Tax" : tax , "Product Type": food_type_detail ,"Sold By" : seller_detail , "Offer" : offer_detail2 , "Feature & Details" :feature_detail
          , "Brand" :brand , "Manufacturer" : manufacurer , 
          "Country of Origin" : country_of_origin , "Disclaimer" : disclaimer  }
```
